```typescript
import { ungzip } from 'npm:node-gzip';
import got from 'npm:got';
import * as xml from "jsr:@libs/xml";

let appstreamUrl = new URL(
  "https://dl.flathub.org/repo/appstream/x86_64/appstream.xml.gz",
);
const { body } = await got(appstreamUrl, {
    responseType: 'buffer',
  });

const xmlString = (await ungzip(body)).toString()

const parsedXml = xml.parse(xmlString)

const appstream = parsedXml.components.component

appstream
```

```typescript

import * as pl from "npm:nodejs-polars@0.18.0";
import * as Plot from "npm:@observablehq/plot";
import { document } from "jsr:@manzt/jupyter-helper";

const x = appstream.map((a)=>{return {
    id:a.id,
    // name:a.name,
    // summary:a.summary,
    categories: typeof a.categories?.category === "string" ? [a.categories?.category]:a.categories?.category,
    keywords: a.keywords && a.keywords?.[0]?.keyword ?
      typeof a.keywords[0].keyword === "string" ? [a.keywords[0].keyword] : a.keywords[0].keyword :
       []
}})

let df: pl.DataFrame = pl.readJSON(JSON.stringify(x))

df = df.select(
  pl.col("id"),
  pl.col("categories"),
  pl.col("keywords"),
);

// let records = df.toRecords();

// Plot.plot({
//   grid: true,
//   inset: 10,
//   aspectRatio: 1,
//   color: {legend: true},
//   marks: [
//     Plot.frame(),
//     Plot.dot(records, {x: "categories", y: "keywords", stroke: "id"})
//   ],
//   document
// })

import { quak } from "jsr:@manzt/jupyter-helper";

quak(df);
```
