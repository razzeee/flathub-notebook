```typescript
import { ungzip } from 'npm:node-gzip';
import got from 'npm:got';
import * as xml from "jsr:@libs/xml";

let appstreamUrl = new URL(
  "https://dl.flathub.org/repo/appstream/x86_64/appstream.xml.gz",
);
const { body } = await got(appstreamUrl, {
    responseType: 'buffer',
  });

const xmlString = (await ungzip(body)).toString()

const parsedXml = xml.parse(xmlString)

const appstream = parsedXml.components.component

console.log(appstream)
```
